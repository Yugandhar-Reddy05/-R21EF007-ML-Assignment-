```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/Users/yugandharreddy/Downloads/archive/adult.csv")
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Education', hue='Income', data=data)
plt.title('Education level vs. Age with Income hue')
plt.xlabel('Age')
plt.ylabel('Education Level')
plt.legend(title='Income')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Income', hue='Gender', data=data)
plt.title('Income Distribution by Gender')
plt.xlabel('Income')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Occupation', hue='Income', data=data)
plt.title('Occupation by Income')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Income')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.countplot(x='Occupation', hue='Income', data=data, dodge=False, palette='Set2')
plt.title('Income Distribution based on Occupation and Gender')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Income')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Income', hue='Gender', data=data)
plt.title('Income by Gender and Age')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend(title='Gender')
plt.show()
```

```python

plt.figure(figsize=(8, 5))
sns.boxplot(x='Income', y='Age', data=data)
plt.title('Age Distribution by Income')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()
```

```python
sns.pairplot(data, hue='Income')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Education', hue='Income', data=data)
plt.title('Income Distribution based on Education')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Income')
plt.show()
```
